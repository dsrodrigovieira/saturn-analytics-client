```python
import pandas as pd
import random

def gerar_dados_hospitalares_completos():
    meses = list(range(1,13))
    hospitais = [9876543, 1234567]
    anos = [2023,2024]
    tipos_atendimento = ["cli","cir"]
    faixas_etarias1 = ["neo_precoce", "neo_tardio", "pedi", "ad", "idoso"]
    faixas_etarias2 = ["neo", "pedi", "ad"]
    tipos_emergencia = ["nivel2", "nivel3"]
    tipos_unidades = ["ui", "uti"]
    tipos_pacientes = ["cli", "cir_orto", "cir_nao_orto"]
    dados = []

    for hospital in hospitais:
        for ano in anos:
            for mes in meses:
                # Indicador 1: Proporção de partos vaginais
                partos_vaginais = random.randint(50, 150)
                partos_cesareos = random.randint(50, 100)

                # Indicador 2: Proporção de reinternações em até 30 dias
                dados_indicador_2 = {}
                for t in tipos_atendimento:
                    saidas_anterior = random.randint(500, 1000)
                    reinternacoes = random.randint(30, 100)
                    dados_indicador_2[f"{t}_saidas_anterior"] = saidas_anterior
                    dados_indicador_2[f"{t}_reinternacoes"] = reinternacoes

                # Indicador 3: Taxa de parada cardiorrespiratória (PCR)
                pacientes_dia = random.randint(4000, 6000)
                pcr_eventos = random.randint(0, 5)

                # Indicador 4: Taxa de mortalidade institucional
                # Indicador 5: Tempo médio de internação
                dados_indicador_4_5 = {}
                for t in tipos_atendimento:
                    for f in faixas_etarias1:
                        obitos = random.randint(5, 20)
                        saidas = random.randint(50, 300)
                        pacientes_dia = random.randint(500, 1000)

                        dados_indicador_4_5[f"{t}_{f}_saidas"] = saidas
                        dados_indicador_4_5[f"{t}_{f}_obitos"] = obitos
                        dados_indicador_4_5[f"{t}_{f}_pacientes_dia"] = pacientes_dia

                # Indicador 6: Tempo médio de permanência na emergência
                total_pacientes_emergencia = random.randint(100, 300)
                total_tempo_permanencia_emergencia_hr = random.randint(300, 900)  # Em horas

                # Indicador 7: Tempo médio de espera na emergência para 1º atendimento
                dados_indicador_7 = {}
                for t in tipos_emergencia:
                    pacientes_emergencia = random.randint(50, 150)
                    tempo_total_emergencia = random.randint(200, 600)  # Em minutos
                    dados_indicador_7[f"pacientes_emergencia_{t}"] = pacientes_emergencia
                    dados_indicador_7[f"tempo_total_emergencia_{t}_min"] = tempo_total_emergencia

                # Indicador 8: Taxa de início de antibiótico intravenoso profilático
                total_cirurgias_limpas = random.randint(50, 150)
                cirurgias_com_antibiotico = random.randint(40, total_cirurgias_limpas)

                # Indicador 9: Taxa de infecção de sítio cirúrgico em cirurgia limpa
                total_cirurgias_limpas_anterior = random.randint(50, 150)
                total_infeccoes = random.randint(0, 5)            

                # Indicador 10: Densidade de incidência de infecção de corrente sanguínea
                dados_indicador_10 = {}
                for u in tipos_unidades:
                    for f in faixas_etarias2:
                        cvc_dia = random.randint(200, 500)
                        infec = random.randint(0, 10)
                        dados_indicador_10[f"{u}_{f}_cvc_dia"] = cvc_dia
                        dados_indicador_10[f"{u}_{f}_infec"] = infec

                # Indicador 11: ITU associada a CVD
                dados_indicador_11 = {}
                for u in tipos_unidades:
                    for f in faixas_etarias2:
                        cvd_dia = random.randint(50, 300)
                        itu_cvd = random.randint(0, 10)
                        dados_indicador_11[f"{u}_{f}_cvd_dia"] = cvd_dia
                        dados_indicador_11[f"{u}_{f}_itu"] = itu_cvd

                # Indicador 12: Profilaxia de TEV
                dados_indicador_12 = {}
                for t in tipos_pacientes:
                    total_pacientes = random.randint(50, 200)
                    profilaxia = random.randint(40, total_pacientes)
                    dados_indicador_12[f"{t}_total_pacientes"] = total_pacientes
                    dados_indicador_12[f"{t}_profilaxia"] = profilaxia

                # Indicador 13: Quedas com lesão
                quedas_com_dano = random.randint(0, 10)

                # Indicador 14: Eventos sentinela
                eventos_sentinela = random.randint(0, 5)

                # Consolidar os dados para o hospital/mês
                dados.append({
                    "cd_cnes": hospital,
                    "ano": ano,
                    "mes": mes,
                    "partos_vaginais": partos_vaginais,
                    "partos_cesareos": partos_cesareos,
                    **dados_indicador_2,
                    "pacientes_dia": pacientes_dia,
                    "pcr_eventos": pcr_eventos,
                    **dados_indicador_4_5,
                    "total_pacientes_emergencia": total_pacientes_emergencia,
                    "total_tempo_permanencia_emergencia_hr": total_tempo_permanencia_emergencia_hr,
                    **dados_indicador_7,
                    "total_cirurgias_limpas": total_cirurgias_limpas,
                    "cirurgias_com_antibiotico": cirurgias_com_antibiotico,
                    "total_cirurgias_limpas_anterior": total_cirurgias_limpas_anterior,
                    "total_infeccoes": total_infeccoes,
                    **dados_indicador_10,
                    **dados_indicador_11,
                    **dados_indicador_12,
                    "quedas_com_dano": quedas_com_dano,
                    "eventos_sentinela": eventos_sentinela
                })

    return pd.DataFrame(dados)

# Gerar os dados e salvar em CSV
dados_completos = gerar_dados_hospitalares_completos()
caminho_arquivo = "dados_hospitalares_completos.csv"
dados_completos.to_csv(caminho_arquivo, index=False)
# print(f"Dados gerados e salvos em: {caminho_arquivo}")
dados_completos
```
